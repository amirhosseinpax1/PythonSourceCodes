```python
# Tester class to manage test cases
class Tester:
    def __init__(self, print_report=True):
        self.print_report = print_report
        self.total_tests = 0
        self.ok_tests = 0
        self.failed_tests = []

    def test_init(self):
        self.total_tests = 0
        self.ok_tests = 0
        self.failed_tests = []

    def test_equal(self, test_name, e1, e2):
        self.total_tests += 1
        if e1 == e2:
            self.ok_tests += 1
        else:
            self.failed_tests.append(test_name)

    def report(self):
        if self.print_report:
            msg = 'OK' if self.ok_tests == self.total_tests else 'FAIL'
            print(f'[{msg}] {self.ok_tests} / {self.total_tests} passed ({self.ok_tests/self.total_tests*100:.2f}%)')
            if self.failed_tests:
                print('Failed tests:', self.failed_tests)
        return self.failed_tests


# Functions to be tested
def isprime(n):
    if n < 2:
        return False
    for i in range(2, n // 2 + 1):
        if n % i == 0:
            return False
    return True

def primes(n):
    return [i for i in range(n + 1) if isprime(i)]


# Test cases for isprime
def test_isprime():
    test = Tester()
    
    test.test_equal('isprime(10)', isprime(10), False)
    test.test_equal('isprime(7)', isprime(7), True)
    test.test_equal('isprime(19)', isprime(19), True)
    test.test_equal('isprime(4)', isprime(4), False)
    test.test_equal('isprime(5)', isprime(5), True)
    test.test_equal('isprime(100)', isprime(100), False)
    test.test_equal('isprime(1)', isprime(1), False)
    test.test_equal('isprime(0)', isprime(0), False)
    test.test_equal('isprime(-3)', isprime(-3), False)

    failed = test.report()
    return failed


# Test cases for primes
def test_primes():
    test = Tester()

    test.test_equal('primes(1)', primes(1), [])
    test.test_equal('primes(7)', primes(7), [2, 3, 5, 7])
    test.test_equal('primes(4)', primes(4), [2, 3])
    test.test_equal('primes(6)', primes(6), [2, 3, 5])
    test.test_equal('primes(10)', primes(10), [2, 3, 5, 7])
    test.test_equal('primes(20)', primes(20), [2, 3, 5, 7, 11, 13, 17, 19])

    failed = test.report()
    return failed


if __name__ == '__main__':
    # Running both sets of tests
    test_isprime()
    test_primes()
```

```python
# کلاس Tester برای مدیریت تست‌ها
class Tester:
    def __init__(self, print_report=True):
        # متغیرها برای ذخیره اطلاعات تست
        self.print_report = print_report  # آیا گزارش تست‌ها چاپ شود یا نه
        self.total_tests = 0  # تعداد کل تست‌ها
        self.ok_tests = 0     # تعداد تست‌های موفق
        self.failed_tests = []  # لیست تست‌های ناموفق

    # متد برای مقداردهی اولیه متغیرهای تست
    def test_init(self):
        self.total_tests = 0
        self.ok_tests = 0
        self.failed_tests = []

    # متد برای مقایسه دو مقدار و ثبت نتیجه تست
    def test_equal(self, test_name, e1, e2):
        self.total_tests += 1  # افزایش تعداد کل تست‌ها
        if e1 == e2:  # اگر دو مقدار برابر باشند
            self.ok_tests += 1  # افزایش تعداد تست‌های موفق
        else:
            self.failed_tests.append(test_name)  # اضافه کردن نام تست به لیست تست‌های ناموفق

    # متد برای چاپ گزارش تست‌ها
    def report(self):
        if self.print_report:  # اگر باید گزارش چاپ شود
            # تعیین وضعیت تست‌ها (موفق یا ناموفق)
            msg = 'OK' if self.ok_tests == self.total_tests else 'FAIL'
            print(f'[{msg}] {self.ok_tests} / {self.total_tests} passed ({self.ok_tests/self.total_tests*100:.2f}%)')
            if self.failed_tests:  # اگر تستی ناموفق بوده است
                print('Failed tests:', self.failed_tests)  # چاپ نام تست‌های ناموفق
        return self.failed_tests  # بازگشت لیست تست‌های ناموفق


# توابعی که باید تست شوند

# تابع برای بررسی اول بودن یک عدد
def isprime(n):
    if n < 2:  # اگر عدد کمتر از 2 باشد
        return False  # عدد اول نیست
    for i in range(2, n // 2 + 1):  # برای هر عدد از 2 تا نصف n
        if n % i == 0:  # اگر n قابل تقسیم بر i باشد
            return False  # عدد اول نیست
    return True  # عدد اول است

# تابع برای تولید لیست اعداد اول تا n
def primes(n):
    return [i for i in range(n + 1) if isprime(i)]  # تولید لیست اعداد اول با استفاده از تابع isprime


# تست‌ها برای تابع isprime
def test_isprime():
    test = Tester()  # ایجاد شیء از کلاس Tester
    
    # تست‌های مختلف برای تابع isprime
    test.test_equal('isprime(10)', isprime(10), False)
    test.test_equal('isprime(7)', isprime(7), True)
    test.test_equal('isprime(19)', isprime(19), True)
    test.test_equal('isprime(4)', isprime(4), False)
    test.test_equal('isprime(5)', isprime(5), True)
    test.test_equal('isprime(100)', isprime(100), False)
    test.test_equal('isprime(1)', isprime(1), False)
    test.test_equal('isprime(0)', isprime(0), False)
    test.test_equal('isprime(-3)', isprime(-3), False)

    failed = test.report()  # گزارش تست‌ها
    return failed  # بازگشت لیست تست‌های ناموفق


# تست‌ها برای تابع primes
def test_primes():
    test = Tester()  # ایجاد شیء از کلاس Tester

    # تست‌های مختلف برای تابع primes
    test.test_equal('primes(1)', primes(1), [])
    test.test_equal('primes(7)', primes(7), [2, 3, 5, 7])
    test.test_equal('primes(4)', primes(4), [2, 3])
    test.test_equal('primes(6)', primes(6), [2, 3, 5])
    test.test_equal('primes(10)', primes(10), [2, 3, 5, 7])
    test.test_equal('primes(20)', primes(20), [2, 3, 5, 7, 11, 13, 17, 19])

    failed = test.report()  # گزارش تست‌ها
    return failed  # بازگشت لیست تست‌های ناموفق


if __name__ == '__main__':
    # اجرای هر دو مجموعه تست
    test_isprime()  # تست تابع isprime
    test_primes()   # تست تابع primes
```
