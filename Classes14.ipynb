```python
import mysql.connector
from mysql.connector import Error

# ایجاد اتصال به MySQL
def create_connection():
    connection = None
    try:
        connection = mysql.connector.connect(
            host="localhost",       # آدرس سرور
            user="Amirhossein",    # نام کاربری
            password="Amir1381",# رمز عبور
        )
        if connection.is_connected():
            print("اتصال به MySQL موفقیت‌آمیز بود")
    except Error as e:
        print(f"خطا در اتصال: {e}")
    
    return connection

# ایجاد دیتابیس
def create_database(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("دیتابیس با موفقیت ایجاد شد")
    except Error as e:
        print(f"خطا در ایجاد دیتابیس: {e}")

# ایجاد جدول
def create_table(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("جدول با موفقیت ایجاد شد")
    except Error as e:
        print(f"خطا در ایجاد جدول: {e}")

# وارد کردن داده به جدول
def insert_data(connection, query, data):
    cursor = connection.cursor()
    try:
        cursor.executemany(query, data)
        connection.commit()  # اعمال تغییرات به دیتابیس
        print("داده‌ها با موفقیت وارد شدند")
    except Error as e:
        print(f"خطا در وارد کردن داده: {e}")

# خواندن داده از جدول
def read_data(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        rows = cursor.fetchall()
        print("اطلاعات موجود در جدول:")
        for row in rows:
            print(row)
    except Error as e:
        print(f"خطا در خواندن داده: {e}")

# مراحل اجرا
def main():
    # 1. ایجاد اتصال
    connection = create_connection()
    
    # 2. ایجاد دیتابیس TestDB
    create_database_query = "CREATE DATABASE IF NOT EXISTS TestDB"
    create_database(connection, create_database_query)
    
    # 3. اتصال به دیتابیس TestDB
    connection.database = "TestDB"
    
    # 4. ایجاد جدول Users
    create_table_query = """
    CREATE TABLE IF NOT EXISTS Users (
        id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(50),
        age INT,
        gender VARCHAR(10)
    )
    """
    
    create_table(connection, create_table_query)
    
    # 5. وارد کردن داده به جدول
    insert_data_query = "INSERT INTO Users (name, age, gender) VALUES (%s, %s, %s)"
    users_data = [
        ("Ali", 25, "Male"),
        ("Sara", 22, "Female"),
        ("Reza", 30, "Male"),
        ("Maryam", 27, "Female")
    ]
    insert_data(connection, insert_data_query, users_data)
    
    # 6. خواندن داده از جدول
    read_data_query = "SELECT * FROM Users Where age > 28"
    read_data(connection, read_data_query)
    
    # 7. بستن اتصال
    if connection.is_connected():
        connection.close()
        print("اتصال بسته شد")

if __name__ == "__main__":
    main()
```

```python
import mysql.connector
from mysql.connector import Error
import tkinter as tk
from tkinter import messagebox

# ایجاد اتصال به MySQL
def create_connection():
    connection = None
    try:
        connection = mysql.connector.connect(
            host="localhost",
            user="Amirhossein",
            password="Amir1381",
            database="TestDB"
        )
        if connection.is_connected():
            print("اتصال به MySQL موفقیت‌آمیز بود")
    except Error as e:
        print(f"خطا در اتصال: {e}")
    
    return connection

# وارد کردن داده به جدول
def insert_data(connection, name, age, gender):
    cursor = connection.cursor()
    try:
        insert_query = "INSERT INTO Users (name, age, gender) VALUES (%s, %s, %s)"
        cursor.execute(insert_query, (name, age, gender))
        connection.commit()
        print("داده‌ها با موفقیت وارد شدند")
    except Error as e:
        print(f"خطا در وارد کردن داده: {e}")

# جستجو کاربران بر اساس نام، جنسیت و محدوده سنی
def search_users(connection, name=None, gender=None, min_age=None, max_age=None, sort_by_age=False):
    cursor = connection.cursor()
    try:
        search_query = "SELECT * FROM Users WHERE 1=1"
        params = []
        
        if name:
            search_query += " AND name = %s"
            params.append(name)
        
        if gender:
            search_query += " AND gender = %s"
            params.append(gender)
        
        if min_age:
            search_query += " AND age >= %s"
            params.append(min_age)
        
        if max_age:
            search_query += " AND age <= %s"
            params.append(max_age)
        
        if sort_by_age:
            search_query += " ORDER BY age"
        
        cursor.execute(search_query, params)
        rows = cursor.fetchall()
        return rows
    except Error as e:
        print(f"خطا در جستجو: {e}")
        return []

# اضافه کردن داده از طریق فرم
def add_user():
    name = name_entry.get()
    age = age_entry.get()
    gender = gender_entry.get()
    
    if name == "" or age == "" or gender == "":
        messagebox.showerror("خطا", "لطفاً تمام فیلدها را پر کنید")
        return
    
    try:
        age = int(age)
    except ValueError:
        messagebox.showerror("خطا", "سن باید یک عدد صحیح باشد")
        return
    
    connection = create_connection()
    insert_data(connection, name, age, gender)
    connection.close()
    messagebox.showinfo("موفقیت", "داده با موفقیت وارد شد")
    clear_entries()

# جستجو کاربران بر اساس ورودی
def search_user():
    name = search_name_entry.get()
    gender = search_gender_entry.get()
    min_age = min_age_entry.get()
    max_age = max_age_entry.get()
    sort_by_age = sort_by_age_var.get()

    try:
        min_age = int(min_age) if min_age else None
        max_age = int(max_age) if max_age else None
    except ValueError:
        messagebox.showerror("خطا", "محدوده سنی باید به درستی وارد شود")
        return
    
    connection = create_connection()
    rows = search_users(connection, name, gender, min_age, max_age, sort_by_age)
    connection.close()

    search_window = tk.Toplevel(window)
    search_window.title("نتایج جستجو")
    
    for i, row in enumerate(rows):
        tk.Label(search_window, text=row).grid(row=i, column=0)

# پاک کردن فیلدهای ورودی
def clear_entries():
    name_entry.delete(0, tk.END)
    age_entry.delete(0, tk.END)
    gender_entry.delete(0, tk.END)
    search_name_entry.delete(0, tk.END)
    search_gender_entry.delete(0, tk.END)
    min_age_entry.delete(0, tk.END)
    max_age_entry.delete(0, tk.END)

# ایجاد رابط کاربری با tkinter
window = tk.Tk()
window.title("برنامه مدیریت کاربران")

# فیلدهای ورود داده
tk.Label(window, text="نام").grid(row=0, column=0)
tk.Label(window, text="سن").grid(row=1, column=0)
tk.Label(window, text="جنسیت").grid(row=2, column=0)

name_entry = tk.Entry(window)
age_entry = tk.Entry(window)
gender_entry = tk.Entry(window)

name_entry.grid(row=0, column=1)
age_entry.grid(row=1, column=1)
gender_entry.grid(row=2, column=1)

# دکمه اضافه کردن کاربر
add_button = tk.Button(window, text="اضافه کردن کاربر", command=add_user)
add_button.grid(row=3, column=0, columnspan=2)

# فیلدهای جستجو
tk.Label(window, text="جستجو بر اساس نام").grid(row=4, column=0)
search_name_entry = tk.Entry(window)
search_name_entry.grid(row=4, column=1)

tk.Label(window, text="جستجو بر اساس جنسیت").grid(row=5, column=0)
search_gender_entry = tk.Entry(window)
search_gender_entry.grid(row=5, column=1)

tk.Label(window, text="حداقل سن").grid(row=6, column=0)
min_age_entry = tk.Entry(window)
min_age_entry.grid(row=6, column=1)

tk.Label(window, text="حداکثر سن").grid(row=7, column=0)
max_age_entry = tk.Entry(window)
max_age_entry.grid(row=7, column=1)

# گزینه مرتب‌سازی بر اساس سن
sort_by_age_var = tk.IntVar()
sort_by_age_check = tk.Checkbutton(window, text="مرتب‌سازی بر اساس سن", variable=sort_by_age_var)
sort_by_age_check.grid(row=8, column=0, columnspan=2)

# دکمه جستجو
search_button = tk.Button(window, text="جستجو", command=search_user)
search_button.grid(row=9, column=0, columnspan=2)

# اجرای پنجره
window.mainloop()
```
