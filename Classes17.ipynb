1. پشته (Stack)
پشته یک ساختار داده است که از اصل LIFO (آخرین ورودی، اولین خروجی) پیروی می‌کند. یعنی عنصری که آخر اضافه شده باشد، اولین عنصر است که حذف می‌شود. پشته را می‌توان با استفاده از لیست پایتون پیاده‌سازی کرد.

```python
class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("Pop from an empty stack")
        return self.stack.pop()

    def peek(self):
        if self.is_empty():
            raise IndexError("Peek from an empty stack")
        return self.stack[-1]

    def size(self):
        return len(self.stack)

# مثال استفاده از پشته
stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)
print(stack.pop())  # خروجی: 30
print(stack.peek())  # خروجی: 20
print(stack.size())  # خروجی: 2
```

صف یک ساختار داده است که از اصل FIFO (اولین ورودی، اولین خروجی) پیروی می‌کند. یعنی عنصری که اول اضافه شده باشد، اولین عنصر است که حذف می‌شود. صف را می‌توان با استفاده از لیست پایتون پیاده‌سازی کرد، اما بهتر است از ماژول collections.deque استفاده کنیم که بهینه‌تر است.

```python
from collections import deque

class Queue:
    def __init__(self):
        self.queue = deque()

    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if self.is_empty():
            raise IndexError("Dequeue from an empty queue")
        return self.queue.popleft()

    def peek(self):
        if self.is_empty():
            raise IndexError("Peek from an empty queue")
        return self.queue[0]

    def size(self):
        return len(self.queue)

# مثال استفاده از صف
queue = Queue()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print(queue.dequeue())  # خروجی: 10
print(queue.peek())     # خروجی: 20
print(queue.size())     # خروجی: 2
```

درخت یک ساختار داده سلسله مراتبی است که از گره‌ها (nodes) تشکیل شده و هر گره می‌تواند چندین فرزند داشته باشد. متداول‌ترین نوع درخت، درخت دودویی است که در آن هر گره حداکثر دو فرزند دارد.

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.value = key

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert(self.root, value)

    def _insert(self, current_node, value):
        if value < current_node.value:
            if current_node.left is None:
                current_node.left = Node(value)
            else:
                self._insert(current_node.left, value)
        else:
            if current_node.right is None:
                current_node.right = Node(value)
            else:
                self._insert(current_node.right, value)

    def inorder_traversal(self, node):
        if node:
            self.inorder_traversal(node.left)
            print(node.value, end=' ')
            self.inorder_traversal(node.right)

# مثال استفاده از درخت دودویی
tree = BinaryTree()
tree.insert(10)
tree.insert(5)
tree.insert(15)
tree.insert(2)
tree.insert(7)

# پیمایش درخت به روش inorder
tree.inorder_traversal(tree.root)  # خروجی: 2 5 7 10 15
```
