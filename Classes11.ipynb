```python
import sqlite3

# اتصال به دیتابیس یا ایجاد یک دیتابیس جدید
connection = sqlite3.connect('my_database.db')

# ایجاد یک cursor برای اجرای دستورات SQL
cursor = connection.cursor()
```

```python
# ایجاد یک جدول به نام 'students'
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    grade TEXT
)
''')

# ذخیره تغییرات
connection.commit()
```

```python
# وارد کردن داده به جدول 'students'
cursor.execute('''
INSERT INTO students (name, age, grade) 
VALUES ('Ali', 20, 'A'),
       ('Sara', 21, 'B'),
       ('Reza', 22, 'C')
''')

# ذخیره تغییرات
connection.commit()
```

```python
# خواندن تمامی داده‌ها از جدول 'students'
cursor.execute('SELECT * FROM students')

# گرفتن نتایج
rows = cursor.fetchall()

# نمایش نتایج
for row in rows:
    print(row)
```

```python
# به‌روزرسانی سن دانشجو با نام 'Ali'
cursor.execute('''
UPDATE students
SET age = 23
WHERE name = 'Ali'
''')

# ذخیره تغییرات
connection.commit()
```

```python
# حذف دانشجو با نام 'Reza'
cursor.execute('''
DELETE FROM students 
WHERE name = 'Reza'
''')

# ذخیره تغییرات
connection.commit()
```

```python
# بستن اتصال به دیتابیس
connection.close()
```

    grade TEXT
);

ALTER TABLE students ADD COLUMN email TEXT;


DROP: برای حذف یک جدول یا پایگاه‌داده.


DROP TABLE students;


TRUNCATE: برای حذف تمامی داده‌های یک جدول بدون حذف خود جدول.

INSERT: برای افزودن داده‌ها به یک جدول.

سناریوی مورد نظر
فرض کنید در یک دانشگاه:

چندین استاد وجود دارند که به دپارتمان‌های مختلف تعلق دارند.
چندین دانشجو وجود دارند که در رشته‌های مختلف تحصیل می‌کنند.
ما باید سیستمی طراحی کنیم که این اطلاعات را در پایگاه‌داده ذخیره و بازیابی کند.

طراحی مدل‌های شی‌گرا
ابتدا کلاس‌های شی‌گرای مربوط به این سناریو را طراحی می‌کنیم:

کلاس Professor: برای نمایندگی یک استاد در سیستم.
کلاس Student: برای نمایندگی یک دانشجو در سیستم.
کلاس University: برای مدیریت استادان و دانشجویان.

```python
class Professor:
    def __init__(self, professor_id, name, department):
        self.professor_id = professor_id
        self.name = name
        self.department = department

    def __str__(self):
        return f'Professor[ID={self.professor_id}, Name={self.name}, Department={self.department}]'


class Student:
    def __init__(self, student_id, name, major):
        self.student_id = student_id
        self.name = name
        self.major = major

    def __str__(self):
        return f'Student[ID={self.student_id}, Name={self.name}, Major={self.major}]'


class University:
    def __init__(self):
        self.professors = []
        self.students = []

    def add_professor(self, professor):
        self.professors.append(professor)
        print(f"Professor '{professor.name}' added to the university.")

    def add_student(self, student):
        self.students.append(student)
        print(f"Student '{student.name}' added to the university.")

    def list_professors(self):
        for professor in self.professors:
            print(professor)

    def list_students(self):
        for student in self.students:
            print(student)






import sqlite3

# ایجاد اتصال به دیتابیس
connection = sqlite3.connect('university.db')
cursor = connection.cursor()

# ایجاد جدول برای استادان
cursor.execute('''
CREATE TABLE IF NOT EXISTS professors (
    professor_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    department TEXT
)
''')

# ایجاد جدول برای دانشجویان
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    major TEXT
)
''')

# ذخیره تغییرات
connection.commit()






def add_professor_to_db(professor):
    cursor.execute('INSERT INTO professors (name, department) VALUES (?, ?)', (professor.name, professor.department))
    connection.commit()
    print(f"Professor '{professor.name}' from {professor.department} added to the database.")



def add_student_to_db(student):
    cursor.execute('INSERT INTO students (name, major) VALUES (?, ?)', (student.name, student.major))
    connection.commit()
    print(f"Student '{student.name}' majoring in {student.major} added to the database.")




# ایجاد نمونه‌هایی از کلاس‌ها
professor1 = Professor(1, 'Dr. Smith', 'Computer Science')
professor2 = Professor(2, 'Dr. Johnson', 'Mathematics')
student1 = Student(1, 'Alice', 'Mathematics')
student2 = Student(2, 'Bob', 'Physics')

# افزودن داده‌ها به پایگاه‌داده
add_professor_to_db(professor1)
add_professor_to_db(professor2)
add_student_to_db(student1)
add_student_to_db(student2)

# ایجاد نمونه‌ای از دانشگاه و افزودن استادان و دانشجویان به آن
university = University()
university.add_professor(professor1)
university.add_professor(professor2)
university.add_student(student1)
university.add_student(student2)

# نمایش لیست استادان و دانشجویان
university.list_professors()
university.list_students()

# بستن اتصال به دیتابیس
connection.close()





```

```python
class Professor:
    def __init__(self, professor_id, name, department):
        self.professor_id = professor_id
        self.name = name
        self.department = department

    def __str__(self):
        return f'Professor[ID={self.professor_id}, Name={self.name}, Department={self.department}]'


class Student:
    def __init__(self, student_id, name, major):
        self.student_id = student_id
        self.name = name
        self.major = major

    def __str__(self):
        return f'Student[ID={self.student_id}, Name={self.name}, Major={self.major}]'



import sqlite3

# ایجاد اتصال به دیتابیس
connection = sqlite3.connect('university.db')
cursor = connection.cursor()

# ایجاد جدول برای استادان
cursor.execute('''
CREATE TABLE IF NOT EXISTS professors (
    professor_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    department TEXT
)
''')

# ایجاد جدول برای دانشجویان
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    major TEXT
)
''')

# ذخیره تغییرات
connection.commit()




def add_professor_to_db(name, department):
    cursor.execute('INSERT INTO professors (name, department) VALUES (?, ?)', (name, department))
    connection.commit()
    print(f"Professor '{name}' from {department} added to the database.")




def add_student_to_db(name, major):
    cursor.execute('INSERT INTO students (name, major) VALUES (?, ?)', (name, major))
    connection.commit()
    print(f"Student '{name}' majoring in {major} added to the database.")




def edit_professor_in_db(professor_id, new_name, new_department):
    cursor.execute('UPDATE professors SET name = ?, department = ? WHERE professor_id = ?', (new_name, new_department, professor_id))
    connection.commit()
    print(f"Professor ID {professor_id} updated in the database.")





def edit_student_in_db(student_id, new_name, new_major):
    cursor.execute('UPDATE students SET name = ?, major = ? WHERE student_id = ?', (new_name, new_major, student_id))
    connection.commit()
    print(f"Student ID {student_id} updated in the database.")





def delete_professor_from_db(professor_id):
    cursor.execute('DELETE FROM professors WHERE professor_id = ?', (professor_id,))
    connection.commit()
    print(f"Professor ID {professor_id} deleted from the database.")




def delete_student_from_db(student_id):
    cursor.execute('DELETE FROM students WHERE student_id = ?', (student_id,))
    connection.commit()
    print(f"Student ID {student_id} deleted from the database.")




def menu():
    while True:
        print("\n--- University Management System ---")
        print("1. Add Professor")
        print("2. Add Student")
        print("3. Edit Professor")
        print("4. Edit Student")
        print("5. Delete Professor")
        print("6. Delete Student")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter professor's name: ")
            department = input("Enter professor's department: ")
            add_professor_to_db(name, department)

        elif choice == '2':
            name = input("Enter student's name: ")
            major = input("Enter student's major: ")
            add_student_to_db(name, major)

        elif choice == '3':
            professor_id = int(input("Enter professor ID to edit: "))
            new_name = input("Enter new name: ")
            new_department = input("Enter new department: ")
            edit_professor_in_db(professor_id, new_name, new_department)

        elif choice == '4':
            student_id = int(input("Enter student ID to edit: "))
            new_name = input("Enter new name: ")
            new_major = input("Enter new major: ")
            edit_student_in_db(student_id, new_name, new_major)

        elif choice == '5':
            professor_id = int(input("Enter professor ID to delete: "))
            delete_professor_from_db(professor_id)

        elif choice == '6':
            student_id = int(input("Enter student ID to delete: "))
            delete_student_from_db(student_id)

        elif choice == '7':
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please try again.")

menu()

# بستن اتصال به دیتابیس
connection.close()
```

انواع داده‌های دیتابیس به دسته‌بندی‌های مختلفی تقسیم می‌شوند و بسته به سیستم مدیریت پایگاه‌داده (مانند SQLite، MySQL، PostgreSQL، SQL Server) ممکن است تفاوت‌های کوچکی در نام‌گذاری یا نوع داده‌ها داشته باشند. در اینجا به معرفی برخی از انواع داده‌های متداول در سیستم‌های مدیریت پایگاه‌داده می‌پردازیم:

1. انواع داده‌های عددی (Numeric Data Types)
این انواع داده‌ها برای ذخیره اعداد (صحیح یا اعشاری) استفاده می‌شوند.

INTEGER: ذخیره اعداد صحیح (بدون اعشار). برای مثال: 1, 42, -5.
REAL: ذخیره اعداد اعشاری (floating-point numbers). برای مثال: 3.14, -0.56.
FLOAT/DOUBLE: مشابه REAL اما با دقت بیشتر برای ذخیره اعداد اعشاری بزرگ.
NUMERIC/DECIMAL: ذخیره اعداد با دقت دقیق اعشاری، به خصوص در امور مالی که دقت مهم است.
2. انواع داده‌های متنی (Text Data Types)
این انواع داده‌ها برای ذخیره داده‌های متنی استفاده می‌شوند.

CHAR(size): ذخیره رشته‌های متنی با طول ثابت. مثلاً CHAR(10) همیشه ۱۰ کاراکتر ذخیره می‌کند.
VARCHAR(size): ذخیره رشته‌های متنی با طول متغیر. مثلاً VARCHAR(50) می‌تواند از ۰ تا ۵۰ کاراکتر ذخیره کند.
TEXT: ذخیره متون طولانی. مثلاً برای توضیحات یا محتوای مقالات.
3. انواع داده‌های باینری (Binary Data Types)
این انواع داده‌ها برای ذخیره داده‌های باینری (مانند تصاویر، ویدئوها و فایل‌ها) استفاده می‌شوند.

BLOB (Binary Large Object): ذخیره داده‌های باینری بزرگ. برای ذخیره فایل‌های تصویری، صوتی، یا هر نوع داده باینری.
4. انواع داده‌های تاریخی و زمانی (Date and Time Data Types)
این انواع داده‌ها برای ذخیره تاریخ و زمان استفاده می‌شوند.

DATE: ذخیره تاریخ. فرمت معمول YYYY-MM-DD. برای مثال: 2024-09-07.
TIME: ذخیره زمان. فرمت معمول HH:MM:SS. برای مثال: 14:30:00.
DATETIME: ذخیره تاریخ و زمان با هم. برای مثال: 2024-09-07 14:30:00.
TIMESTAMP: ذخیره زمان در قالبی که همواره به‌روزرسانی شود (بسته به تنظیمات). معمولاً برای زمان‌های ثبت تغییرات.
YEAR: ذخیره سال. معمولاً دو یا چهار رقمی. برای مثال: 2024 یا 24.
5. انواع داده‌های منطقی (Boolean Data Types)
این نوع داده برای ذخیره مقادیر منطقی استفاده می‌شود.

BOOLEAN: ذخیره مقادیر TRUE یا FALSE (در بسیاری از پایگاه‌های داده، به‌عنوان 1 و 0 ذخیره می‌شوند).
6. انواع داده‌های ویژه (Special Data Types)
این انواع داده‌ها برای کاربردهای خاص طراحی شده‌اند.

ENUM: ذخیره مجموعه‌ای از مقادیر از پیش تعریف شده. برای مثال: ENUM('Male', 'Female', 'Other').
SET: ذخیره مجموعه‌ای از مقادیر چندگانه از پیش تعریف شده. برای مثال: SET('Reading', 'Writing', 'Coding').
7. انواع داده‌های جغرافیایی (Geospatial Data Types)
این نوع داده‌ها برای ذخیره اطلاعات جغرافیایی مانند مختصات، مکان‌ها، و داده‌های نقشه‌برداری استفاده می‌شوند.

GEOMETRY: ذخیره داده‌های هندسی عمومی.
POINT: ذخیره یک نقطه جغرافیایی (X, Y).
LINESTRING: ذخیره یک خط چند نقطه‌ای.
POLYGON: ذخیره یک چندضلعی.
8. انواع داده‌های JSON و XML
این نوع داده‌ها برای ذخیره داده‌های ساختار یافته مانند اسناد JSON یا XML استفاده می‌شوند.

JSON: ذخیره داده‌ها در قالب JSON.
XML: ذخیره داده‌ها در قالب XML.
