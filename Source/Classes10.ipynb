توابع tell و seek و نقش آن‌ها در خواندن محتوای فایل
tell() برای نشان دادن مکان فعلی در فایل و seek() برای حرکت به موقعیت مشخصی در فایل استفاده می‌شوند.

```python
with open("example.txt", "r") as file:
    print("Current position:", file.tell())
    print(file.read(5))  # خواندن پنج کاراکتر
    print("Current position:" , file.tell())
    file.seek(0)  # بازگشت به ابتدای فایل
    print(file.read(5))  # خواندن پنج کاراکتر اول دوباره
```

فایلی ایجاد کنید که شامل ۵ خط متن باشد و سپس هر خط را بخوانید و چاپ کنید.


```python
with open("example.txt", "w") as file:
    for i in range(5):
        file.write(f"Line {i+1}\n")

with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())
```

فایلی بنویسید که نام کاربری و کلمه عبور کاربران را در یک فایل ذخیره کند. سپس این فایل را باز کرده و داده‌ها را چاپ کنید.


```python
with open("users.txt", "w") as file:
    file.write("user1:password1\n")
    file.write("user2:password2\n")

with open("users.txt", "r") as file:
    print(file.read())
```

فایل‌های متنی بزرگ را خط به خط بخوانید و تعداد خطوطی که شامل کلمه‌ای خاص هستند را بشمارید.


```python
keyword = "specific_word"
count = 0
with open("large_file.txt", "r") as file:
    for line in file:
        if keyword in line:
            count += 1
print(f"Number of lines with '{keyword}':", count)
```

فایلی ایجاد کنید که هر خط از آن شامل یک عدد است. سپس مجموع این اعداد را محاسبه کنید.


```python
with open("numbers.txt", "w") as file:
    file.write("10\n20\n30\n40\n50\n")

total = 0
with open("numbers.txt", "r") as file:
    for line in file:
        total += int(line.strip())
print("Sum of numbers:", total)
```

در پایتون، اگر بخواهید به محتوای یک فایل چیزی اضافه کنید بدون اینکه محتوای قبلی آن فایل حذف شود، باید از حالت append یا a استفاده کنید. در این حالت، اگر فایل وجود داشته باشد، به انتهای آن محتوای جدید اضافه می‌شود؛ و اگر فایل وجود نداشته باشد، فایلی جدید با همان نام ایجاد شده و محتوا در آن نوشته می‌شود.

```python
# فایل را در حالت اضافه کردن باز کنید
file = open("example.txt", "a")

# اضافه کردن خط جدید
file.write("This is an appended line.\n")

# بستن فایل
file.close()

# خواندن و چاپ محتوای جدید فایل
with open("example.txt", "r") as file:
    print(file.read())
```

```python
# فایل را در حالت اضافه کردن باز کنید
with open("example.txt", "a") as file:
    lines = ["First appended line.\n", "Second appended line.\n", "Third appended line.\n"]
    file.writelines(lines)

# خواندن و چاپ محتوای جدید فایل
with open("example.txt", "r") as file:
    print(file.read())
```

تمرین: فایلی ایجاد کنید و یک لیست از نام‌ها را در آن بنویسید. سپس برنامه‌ای بنویسید که دو نام جدید به انتهای این فایل اضافه کند و محتوای کامل فایل را چاپ کند.

```python
# مرحله اول: ایجاد فایل و نوشتن نام‌ها
with open("names.txt", "w") as file:
    names = ["ali\n", "reza\n", "Mmd\n"]
    file.writelines(names)

# مرحله دوم: اضافه کردن نام‌های جدید
with open("names.txt", "a") as file:
    new_names = ["sara\n", "yalda\n"]
    file.writelines(new_names)

# خواندن و چاپ محتوای کامل فایل
with open("names.txt", "r") as file:
    print(file.read())
```

CSV (Comma-Separated Values) یک فرمت رایج برای ذخیره داده‌ها در قالب متنی است که در آن مقادیر داده‌ها با کاما (یا دیگر جداکننده‌ها) از هم جدا می‌شوند. هر خط یک رکورد داده است.

```python
#name,age,city
#Ali,25,Tehran
#Sara,30,Isfahan



import csv

data = [
    ['name', 'age', 'city'],
    ['Ali', '25', 'Tehran'],
    ['Sara', '30', 'Isfahan']
]

with open('output.csv', mode='w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)


```

```python
    
import csv

with open('output.csv', mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

JSON (JavaScript Object Notation) یک فرمت استاندارد برای تبادل داده است که در آن داده‌ها به صورت جفت کلید-مقدار نوشته می‌شوند.

```python
# {
#     "name": "Ali",
#     "age": 25,
#     "city": "Tehran"
# }


import json

data = {
    "name": "Sara",
    "age": 30,
    "city": "Isfahan"
}

with open('output.json', mode='w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=40)
```

```python
import json

with open('output.json', mode='r', encoding='utf-8') as file:
    data = json.load(file)
    print(data)
```

XML (eXtensible Markup Language) یک فرمت نشانه‌گذاری برای ساختاردهی و تبادل داده‌ها است.



```python
# <person>
#     <name>Ali</name>
#     <age>25</age>
#     <city>Tehran</city>
# </person>


import xml.etree.ElementTree as ET

person = ET.Element('person')
name = ET.SubElement(person, 'name')
name.text = 'Sara'
age = ET.SubElement(person, 'age')
age.text = '30'
city = ET.SubElement(person, 'city')
city.text = 'Isfahan'

tree = ET.ElementTree(person)
tree.write('output.xml', encoding='utf-8', xml_declaration=True)
```

```python
import xml.etree.ElementTree as ET

tree = ET.parse('output.xml')
root = tree.getroot()

for child in root:
    print(child.tag, child.text)
```
