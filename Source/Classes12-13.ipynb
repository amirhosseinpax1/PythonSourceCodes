```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import sqlite3

# ایجاد یا اتصال به دیتابیس SQLite
conn = sqlite3.connect('students.db')
c = conn.cursor()

# ایجاد جدول در دیتابیس (در صورت عدم وجود)
c.execute('''
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    major TEXT NOT NULL
)
''')
conn.commit()

# تابع برای ذخیره کردن دانشجویان در دیتابیس
def save_student():
    name = entry_name.get()
    major = entry_major.get()
    
    if name and major:
        c.execute("INSERT INTO students (name, major) VALUES (?, ?)", (name, major))
        conn.commit()
        messagebox.showinfo("Success", "Student added successfully!")
        entry_name.delete(0, tk.END)
        entry_major.delete(0, tk.END)
    else:
        messagebox.showwarning("Input error", "Please fill both fields")

# تابع برای باز کردن پنجره‌ی نمایش لیست دانشجویان
def show_students():
    # ایجاد یک پنجره‌ی جدید
    root2 = tk.Toplevel(root)
    root2.title("Student List")
    

 

    # ایجاد جدول
    tree = ttk.Treeview(root2, columns=("ID", "Name", "Major"), show="headings")
    tree.heading("ID", text="آیدی")
    tree.heading("Name", text="نام")
    tree.heading("Major", text="رشته")
    
    tree.pack(fill=tk.BOTH, expand=False)

    # واکشی اطلاعات از دیتابیس و پر کردن جدول
    c.execute("SELECT * FROM students")
    rows = c.fetchall()
    
    for x in rows:
        tree.insert("", tk.END, values=x)

# رابط کاربری اصلی
root = tk.Tk()
root.title("Student Management")

# فریم برای فرم ورود اطلاعات
frame = tk.Frame(root)
frame.pack(pady=20)

tk.Label(frame, text="Student Name").grid(row=0, column=0, padx=10, pady=10)
tk.Label(frame, text="Major").grid(row=1, column=0, padx=10, pady=10)

entry_name = tk.Entry(frame)
entry_name.grid(row=0, column=1, padx=10, pady=10)

entry_major = tk.Entry(frame)
entry_major.grid(row=1, column=1, padx=10, pady=10)

btn_save = tk.Button(frame, text="Save Student", command=save_student)
btn_save.grid(row=2, columnspan=2, pady=20)

btn_show = tk.Button(root, text="Show Students", command=show_students)
btn_show.pack(pady=10)

root.mainloop()

# بستن اتصال به دیتابیس در انتهای برنامه
conn.close()
```

```python
import tkinter as tk  # وارد کردن کتابخانه tkinter برای ساخت رابط کاربری گرافیکی

# تابع برای مدیریت کلیک دکمه‌ها (برای اعداد و عملگرها)
def button_click(number):
    current = entry.get()  # گرفتن متن فعلی از ویجت ورودی
    entry.delete(0, tk.END)  # پاک کردن ویجت ورودی
    entry.insert(0, str(current) + str(number))  # درج متن جدید (عدد یا عملگر کلیک شده) در ورودی

# این تابع زمانی فراخوانی می‌شود که یک دکمه (عددی یا عملگری) کلیک شود. ابتدا متن فعلی ورودی را می‌گیرد، سپس آن را پاک کرده و متن جدید (عدد یا عملگر) را اضافه می‌کند.


# تابع برای پاک کردن محتوای ورودی
def button_clear():
    entry.delete(0, tk.END)  # پاک کردن تمام محتوای ویجت ورودی

# تابع برای محاسبه نتیجه و نمایش آن
def button_equal():
    try:
        result = str(eval(entry.get()))  # ارزیابی و محاسبه عبارت ریاضی از ورودی
        entry.delete(0, tk.END)  # پاک کردن ورودی قبلی
        entry.insert(0, result)  # نمایش نتیجه در ویجت ورودی
    except:
        entry.delete(0, tk.END)  # در صورت خطا، ورودی را پاک کن
        entry.insert(0, "خطا!")  # نمایش پیغام "خطا!" در صورت وجود مشکل


# این تابع زمانی فراخوانی می‌شود که دکمه = کلیک شود. ابتدا سعی می‌کند با استفاده از تابع eval محتوای ورودی را به عنوان یک عبارت ریاضی ارزیابی کند. اگر محاسبه موفقیت‌آمیز بود، نتیجه نمایش داده می‌شود، وگرنه پیغام "خطا!" نمایش داده می‌شود.



root = tk.Tk()  # ساخت پنجره اصلی برنامه با استفاده از tkinter
root.title("ماشین حساب")  # تنظیم عنوان پنجره به "ماشین حساب"

# ساخت یک ویجت ورودی برای نمایش و ورود اعداد
entry = tk.Entry(root, width=35, borderwidth=5)
entry.grid(row=0, column=0, columnspan=4)  # قرار دادن ویجت ورودی در شبکه (grid)

# لیستی از دکمه‌های ماشین حساب (اعداد و عملگرها)
buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+'
]

row_val = 1  # شماره سطر برای دکمه‌ها
col_val = 0  # شماره ستون برای دکمه‌ها

# ساخت و قرار دادن دکمه‌ها در پنجره
for button in buttons:
    action = lambda x=button: button_click(x) if x != '=' else button_equal()  # تعیین عملکرد دکمه (اگر '=' باشد، تابع button_equal را فراخوانی کن، در غیر این صورت button_click)
    tk.Button(root, text=button, width=9, height=3, command=action).grid(row=row_val, column=col_val)  # ساخت دکمه و قرار دادن آن در شبکه
    col_val += 1  # حرکت به ستون بعدی
    if col_val > 3:  # اگر ستون‌ها پر شد، به سطر بعدی برو
        col_val = 0  # تنظیم شماره ستون به 0
        row_val += 1  # افزایش شماره سطر

# ساخت دکمه "پاک کردن" و قرار دادن آن در پنجره
tk.Button(root, text="پاک کردن", width=9, height=3, command=button_clear).grid(row=row_val, columnspan=4)

root.mainloop()  # اجرای حلقه اصلی برنامه برای نمایش پنجره و مدیریت رویدادها
```

توضیحات تکمیلی 

1. Entry(cal, ...):
Entry: این یک ویجت (ابزارک) از کتابخانه Tkinter است که برای نمایش و دریافت ورودی متنی در رابط کاربری گرافیکی استفاده می‌شود. در اینجا برای نمایش اعداد و نتایج محاسبات در ماشین حساب به کار می‌رود.
cal: این اولین آرگومان، نام پنجره یا فریم والد (در اینجا cal که شیء اصلی پنجره است) را مشخص می‌کند که این فیلد ورودی در آن قرار می‌گیرد.
2. font=('arial', 20, 'bold'):
font: مشخصات فونت متن داخل فیلد ورودی را تعیین می‌کند.
'arial': نام فونت را مشخص می‌کند (در اینجا فونت Arial).
20: اندازه فونت را مشخص می‌کند (در اینجا اندازه فونت 20 است).
'bold': استایل فونت را مشخص می‌کند (در اینجا فونت به صورت ضخیم یا Bold است).
3. textvariable=text_input:
textvariable: این پارامتر تعیین می‌کند که محتوای فیلد ورودی با یک متغیر در ارتباط باشد.
text_input: این متغیری از نوع StringVar() است که محتوای فیلد ورودی را ذخیره و مدیریت می‌کند. با استفاده از این متغیر، می‌توان مقدار فیلد ورودی را به راحتی تغییر داد یا خواند.
4. bd=30:
bd (Border Width): این پارامتر تعیین می‌کند که عرض حاشیه (border) اطراف فیلد ورودی چند پیکسل باشد. در اینجا حاشیه 30 پیکسل است که به وضوح باعث بزرگتر شدن کادر نمایشگر می‌شود.
5. insertwidth=4:
insertwidth: این پارامتر عرض نشانگر چشمک‌زن (caret) که در هنگام وارد کردن یا نمایش متن ظاهر می‌شود را تنظیم می‌کند. مقدار 4 به این معنی است که عرض نشانگر 4 پیکسل خواهد بود.
6. bg='powder blue':
bg (Background): این پارامتر رنگ پس‌زمینه فیلد ورودی را تعیین می‌کند. در اینجا از رنگ powder blue (آبی کم‌رنگ) برای پس‌زمینه استفاده شده است.
7. justify='right':
justify: نحوه تراز کردن متن داخل فیلد ورودی را مشخص می‌کند. مقدار 'right' به این معنی است که متن از سمت راست فیلد ورودی تراز می‌شود. این ویژگی مناسب نمایش اعداد در ماشین حساب است، زیرا معمولاً اعداد در ماشین حساب‌ها از سمت راست وارد می‌شوند.
8. .grid(columnspan=4):
grid(): این یک متد برای قرار دادن ویجت‌ها در رابط کاربری با استفاده از یک سیستم شبکه‌ای (grid) است. در اینجا از grid() برای تنظیم موقعیت فیلد ورودی در رابط کاربری استفاده می‌شود.
columnspan=4: این پارامتر مشخص می‌کند که این فیلد ورودی باید به اندازه ۴ ستون در رابط کاربری گسترش یابد. چون ماشین حساب در کل ۴ ستون دارد، این ویژگی باعث می‌شود که فیلد ورودی در بالای ماشین حساب به صورت کامل (در تمام ۴ ستون) نمایش داده شود.

sticky="nsew":
پارامتر sticky به ما این امکان را می‌دهد که دکمه‌ها را در جهات مختلف بکشیم. در مثال بالا، با استفاده از sticky="nsew"، دکمه 2 به تمام فضای شبکه (north, south, east, west) کشیده می‌شود.

```python
import tkinter as tk

root = tk.Tk()
root.geometry("600x300")

# ایجاد دکمه‌ها
btn1 = tk.Button(root, text="Button 1", padx=20, pady=10)
btn2 = tk.Button(root, text="Button 2", padx=20, pady=10)
btn3 = tk.Button(root, text="Button 3", padx=20, pady=10)

# قرار دادن دکمه‌ها در شبکه (grid)
btn1.grid(row=0, column=0)  # دکمه 1 در ستون 0 و ردیف 0
btn2.grid(row=0, column=1, columnspan=2, sticky="nsew")  # دکمه 2 دو ستون را اشغال می‌کند
btn3.grid(row=0, column=3)  # دکمه 3 در ستون 3

# تنظیم کشش ستون‌ها
root.grid_columnconfigure(0, weight=1)
root.grid_columnconfigure(1, weight=1)
root.grid_columnconfigure(2, weight=1)
root.grid_columnconfigure(3, weight=1)

root.mainloop()



```

```python
import tkinter as tk

# ایجاد پنجره اصلی
root = tk.Tk()
root.title("ماشین حساب زیبا")
root.geometry("400x600")
root.config(bg="#1e1e1e")

# استایل نمایشگر
display = tk.Entry(root, font=("Helvetica", 32), bg="#1e1e1e", fg="#ffffff", bd=0, justify="right", insertbackground="#ffffff")
display.grid(row=0, column=0, columnspan=4, padx=10, pady=20, ipady=10)

# استایل دکمه‌ها
button_font = ("Helvetica", 18)
button_color = "#2e2e2e"
button_text_color = "#ffffff"
button_active_color = "#4e4e4e"

# تابع برای نمایش اعداد
def button_click(number):
    current = display.get()
    display.delete(0, tk.END)
    display.insert(0, current + str(number))

# ایجاد دکمه‌های اعداد و عملیات
button_texts = [
    ('7', 2, 0), ('8', 2, 1), ('9', 2, 2), ('+', 2, 3),
    ('4', 3, 0), ('5', 3, 1), ('6', 3, 2), ('-', 3, 3),
    ('1', 4, 0), ('2', 4, 1), ('3', 4, 2), ('*', 4, 3),
    ('0', 5, 0, 2), ('.', 5, 2), ('/', 5, 3)
]

for (text, row, col, *span) in button_texts:
    span = span[0] if span else 1
    btn = tk.Button(root, text=text, font=button_font, bg=button_color, fg=button_text_color, activebackground=button_active_color, bd=0, padx=20, pady=20)
    btn.config(command=lambda t=text: button_click(t))
    btn.grid(row=row, column=col, columnspan=span, sticky="nsew", padx=5, pady=5)

root.mainloop()
```

```python
import tkinter as tk

# ایجاد پنجره اصلی
root = tk.Tk()
root.title("ماشین حساب زیبا")
root.geometry("400x600")
root.config(bg="#1e1e1e")


root.iconbitmap("images/a.ico")  # تغییر آیکون برنامه با فایل icon.ico


# استایل نمایشگر
display = tk.Entry(root, font=("Helvetica", 32), bg="#1e1e1e", fg="#ffffff", bd=0, justify="right", insertbackground="#ffffff")
display.grid(row=0, column=0, columnspan=4, padx=10, pady=20, ipady=10)

# استایل دکمه‌ها
button_font = ("Helvetica", 18)
button_color = "#2e2e2e"
button_text_color = "#ffffff"
button_active_color = "#4e4e4e"

# تابع برای نمایش اعداد
def button_click(number):
    current = display.get()
    display.delete(0, tk.END)
    display.insert(0, current + str(number))

# تابع پاک کردن نمایشگر
def button_clear():
    display.delete(0, tk.END)

# تابع محاسبه نتیجه
def button_equal():
    try:
        result = str(eval(display.get()))  # محاسبه و نمایش نتیجه
        display.delete(0, tk.END)
        display.insert(0, result)
    except:
        display.delete(0, tk.END)
        display.insert(0, "خطا")

# ایجاد دکمه‌های اعداد و عملیات
button_texts = [
    ('C', 1, 0), ('/', 1, 1), ('*', 1, 2), ('-', 1, 3),
    ('7', 2, 0), ('8', 2, 1), ('9', 2, 2), ('+', 2, 3),
    ('4', 3, 0), ('5', 3, 1), ('6', 3, 2), ('=', 3, 3 ),
    ('1', 4, 0), ('2', 4, 1), ('3', 4, 2),
    ('0', 5, 0, 2), ('.', 5, 2),
]

for (text, row, col, *span) in button_texts:
    span = span[0] if span else 1
    btn = tk.Button(root, text=text, font=button_font, bg=button_color, fg=button_text_color, activebackground=button_active_color, bd=0, padx=20, pady=20)
    
    if text.isdigit() or text == ".":
        btn.config(command=lambda t=text: button_click(t))
    elif text == 'C':
        btn.config(command=button_clear)
    elif text == '=':
        btn.config(command=button_equal)
    else:
        btn.config(command=lambda t=text: button_click(t))
    
    btn.grid(row=row, column=col, columnspan=span, sticky="nsew", padx=5, pady=5)

# تنظیم کشش ستون‌ها و ردیف‌ها
for i in range(4):
    root.grid_columnconfigure(i, weight=1)  # کشش ستون‌ها
for i in range(6):
    root.grid_rowconfigure(i, weight=1)  # کشش ردیف‌ها

root.mainloop()
```

```python
import tkinter as tk

# ایجاد پنجره اصلی
# با استفاده از Tkinter یک پنجره اصلی به نام root ساخته می‌شود
root = tk.Tk()
root.title("ماشین حساب زیبا")  # عنوان پنجره
root.geometry("400x600")  # تنظیم اندازه پنجره به عرض 400 و ارتفاع 600
root.config(bg="#1e1e1e")  # تنظیم رنگ پس‌زمینه پنجره به رنگ تیره (برای ظاهری مدرن)

# استایل نمایشگر
# نمایشگر ماشین‌حساب که ورودی‌ها و خروجی‌ها در آن نمایش داده می‌شود (Entry widget)
display = tk.Entry(root, font=("Helvetica", 32), bg="#1e1e1e", fg="#ffffff", bd=0, justify="right", insertbackground="red")
# font: فونت و اندازه متن
# bg: رنگ پس‌زمینه نمایشگر
# fg: رنگ متن (عدد)
# bd: بدون border
# justify: تراز کردن متن به سمت راست
# insertbackground: رنگ نشانگر هنگام تایپ
display.grid(row=0, column=0, columnspan=4, padx=100, pady=20, ipady=10)
# grid: قرار دادن نمایشگر در پنجره با استفاده از سیستم grid
# row: ردیف قرارگیری
# column: ستون قرارگیری
# columnspan: تعداد ستون‌هایی که نمایشگر اشغال می‌کند (4 ستون)
# padx و pady: فاصله‌های داخلی از اطراف
# ipady: افزایش ارتفاع داخلی نمایشگر

# استایل دکمه‌ها
button_font = ("Helvetica", 18)  # تنظیم فونت و اندازه متن دکمه‌ها
button_color = "#2e2e2e"  # رنگ پس‌زمینه دکمه‌ها
button_text_color = "#ffffff"  # رنگ متن دکمه‌ها
button_active_color = "#4e4e4e"  # رنگ پس‌زمینه دکمه‌ها در حالت فعال (زمانی که فشرده می‌شوند)

# تابع برای نمایش اعداد در نمایشگر
def button_click(number):
    current = display.get()  # دریافت مقدار فعلی از نمایشگر
    display.delete(0, tk.END)  # پاک کردن مقدار قبلی در نمایشگر
    display.insert(0, current + str(number))  # افزودن عدد جدید به انتهای مقدار قبلی

# تابع برای پاک کردن نمایشگر
def button_clear():
    display.delete(0, tk.END)  # پاک کردن تمامی متن از نمایشگر

# تابع محاسبه نتیجه
def button_equal():
    try:
        result = str(eval(display.get()))  # ارزیابی عبارت ریاضی وارد شده توسط کاربر و تبدیل آن به رشته
        display.delete(0, tk.END)  # پاک کردن مقدار قبلی
        display.insert(0, result)  # نمایش نتیجه محاسبه
    except:
        display.delete(0, tk.END)  # در صورت بروز خطا، پاک کردن نمایشگر
        display.insert(0, "خطا")  # نمایش پیام خطا در نمایشگر

# ایجاد دکمه‌های اعداد و عملیات ریاضی
# یک لیست از دکمه‌های مختلف با متن، ردیف و ستون مربوطه ساخته شده
button_texts = [
    ('C', 1, 0), ('/', 1, 1), ('*', 1, 2), ('-', 1, 3),
    ('7', 2, 0), ('8', 2, 1), ('9', 2, 2), ('+', 2, 3),
    ('4', 3, 0), ('5', 3, 1), ('6', 3, 2), ('=', 3, 3 ),
    ('1', 4, 0), ('2', 4, 1), ('3', 4, 2),
    ('0', 5, 0, 2), ('.', 5, 2),
]

# این حلقه برای ایجاد و تنظیم دکمه‌های ماشین‌حساب است
for (text, row, col, *span) in button_texts:
    span = span[0] if span else 1  # اگر span تعیین شده بود، آن را می‌گیرد وگرنه 1 (یعنی یک ستون)
    # ایجاد دکمه با متن و تنظیمات مشخص
    btn = tk.Button(root, text=text, font=button_font, bg=button_color, fg=button_text_color, activebackground=button_active_color, bd=0, padx=20, pady=20)
    
    # اگر دکمه عدد یا '.' باشد، تابع button_click برای کلیک شدن تنظیم می‌شود
    if text.isdigit() or text == ".":
        btn.config(command=lambda t=text: button_click(t))
    # اگر دکمه 'C' باشد، تابع button_clear برای پاک کردن استفاده می‌شود
    elif text == 'C':
        btn.config(command=button_clear)
    # اگر دکمه '=' باشد، تابع button_equal برای محاسبه نتیجه استفاده می‌شود
    elif text == '=':
        btn.config(command=button_equal)
    # سایر دکمه‌ها (مانند عملیات ریاضی) هم به تابع button_click متصل می‌شوند
    else:
        btn.config(command=lambda t=text: button_click(t))
    
    # قرار دادن دکمه در پنجره با استفاده از grid
    btn.grid(row=row, column=col, columnspan=span, sticky="nsew", padx=5, pady=5)

# تنظیم کشش ستون‌ها و ردیف‌ها
for i in range(4):
    root.grid_columnconfigure(i, weight=1)  # تنظیم ستون‌ها برای کشش یکنواخت و پر کردن فضا
for i in range(6):
    root.grid_rowconfigure(i, weight=1)  # تنظیم ردیف‌ها برای کشش یکنواخت و پر کردن فضا

# اجرای اصلی برنامه
root.mainloop()  # شروع اجرای برنامه و نمایش پنجره تا زمان بسته شدن
```
