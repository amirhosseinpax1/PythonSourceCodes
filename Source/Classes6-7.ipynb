A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

```python
x = lambda a : a + 10
print(x(5)) 
```

```python
x = lambda a, b : a * b
print(x(5, 6)) 
```

```python
x = lambda a, b, c : a + b + c
print(x(5, 6, 2)) 
```

#lambda
#توابع بی‌نام هستند که به صورت فشرده تعریف می‌شوند. این توابع معمولاً برای عملیات‌های ساده‌ای که نیازی به تعریف کامل تابع ندارند، استفاده می‌شوند.

```python
square = lambda x: x ** 2
print(square(5))  # Output: 25
```

```python
add = lambda a, b: a + b
print(add(3, 4))  # Output: 7
```

```python
# تابع 
# map 
# در پایتون برای اعمال یک تابع مشخص به تمامی آیتم‌های یک
# iterable 
# (مانند لیست یا تاپل) استفاده می‌شود. 
# این تابع یک شیء برمی‌گرداند که می‌تواند به لیست یا نوع دیگری تبدیل شود.

#map(function, iterable)

numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]

```

```python
strings = ["1", "2", "3"]
integers = map(int, strings)
print(list(integers))  # Output: [1, 2, 3]
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
summed = map(lambda x, y: x + y, list1, list2)
print(list(summed))  # Output: [5, 7, 9]
```

```python
# تابع 
# zip 
# در پایتون برای ترکیب چندین
#  iterable 
# به یک
#  iterable 
# جدید که شامل تاپل‌های عناصر متناظر از هر
#  iterable 
# است، استفاده می‌شود.


#zip(iterable1, iterable2, ...)


names = ['Alice', 'Bob', 'Charlie']
scores = [85, 90, 95]
combined = zip(names, scores)
print(list(combined))  # Output: [('Alice', 85), ('Bob', 90), ('Charlie', 95)]

```

```python
names = ['Alice', 'Bob']
scores = [85, 90]
ages = [20, 21]
combined = zip(names, scores, ages)
print(list(combined))  # Output: [('Alice', 85, 20), ('Bob', 90, 21)]
```

```python
letters = ['a', 'b', 'c']
indexed = zip(range(len(letters)), letters)
print(list(indexed))  # Output: [(0, 'a'), (1, 'b'), (2, 'c')]
```

```python
# تابع
#  filter 
# در پایتون برای فیلتر کردن عناصر یک
#  iterable 
# با توجه به تابعی که شرطی را بررسی می‌کند، استفاده می‌شود. اگر شرط برقرار باشد، عنصر در نتیجه قرار می‌گیرد.


#filter(function, iterable)

numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4, 6]


```

```python
words = ["apple", "cat", "dog", "banana"]
short_words = filter(lambda word: len(word) < 4, words)
print(list(short_words))  # Output: ['cat', 'dog']
```

```python
numbers = [-10, -5, 0, 5, 10]
positive_numbers = filter(lambda x: x > 0, numbers)
print(list(positive_numbers))  # Output: [5, 10]
```

تمرین 1: استفاده از lambda و map
یک لیست از اعداد صحیح داده شده است. یک تابع lambda بنویسید که هر عدد را به توان 3 برساند و از map برای اعمال این تابع بر روی لیست استفاده کنید.

```python
numbers = [1, 2, 3, 4, 5]
cubed_numbers = list(map(lambda x: x ** 3, numbers))
print(cubed_numbers)  # Output: [1, 8, 27, 64, 125]
```

تمرین 2: استفاده از filter و lambda
یک لیست از رشته‌ها داده شده است. با استفاده از filter و lambda، فقط رشته‌هایی که طولشان بیشتر از 4 کاراکتر است را فیلتر کنید.

```python
words = ["apple", "cat", "banana", "dog", "elephant"]
filtered_words = list(filter(lambda word: len(word) > 4, words))
print(filtered_words)  # Output: ['apple', 'banana', 'elephant']
```

تمرین 3: استفاده از zip
دو لیست از نام‌ها و نمرات دانش‌آموزان داده شده است. با استفاده از zip، لیستی از تاپل‌ها ایجاد کنید که هر تاپل شامل نام دانش‌آموز و نمره‌ی او باشد.

```python
names = ["َAli", "Reza", "Mohammad"]
scores = [85, 90, 78]
student_scores = list(zip(names, scores))
print(student_scores)  # Output: [('َAli', 85), ('Reza', 90), ('Mohammad', 78)]
```
