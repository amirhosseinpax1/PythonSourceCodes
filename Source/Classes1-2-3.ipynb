```python
# Exercise: Determine if three sides can form a triangle
# تمرین: بررسی اینکه آیا سه ضلع می‌توانند مثلثی تشکیل دهند یا خیر
a = int(input("Enter side 1: "))
b = int(input("Enter side 2: "))
c = int(input("Enter side 3: "))


if (a + b > c) and (a + c > b) and (b + c > a):
    print("These sides can form a triangle.")
else:
    print("These sides cannot form a triangle.")



# تمرین: بررسی نوع مثلث (متساوی‌الاضلاع، متساوی‌الساقین، یا مختلف‌الاضلاع)
if a == b == c:
    print("Equilateral triangle")
elif a == b or b == c or a == c:
    print("Isosceles triangle")
else:
    print("Scalene triangle")
```

```python
# Exercise: Count and display the number of vowels in a string
# تمرین: شمارش و نمایش تعداد حروف صدادار در یک رشته
s = input("Enter a string: ")
vowels = "AaEeIiOoUu"
count = 0
for i in s:
    if i in vowels:
        count += 1
print(f"Number of vowels: {count}")



# تمرین: چاپ حروف بی‌صدا به همراه تعداد آنها
consonants_count = len([i for i in s if i not in vowels])
print(f"Number of consonants: {consonants_count}")
```

```python
# Exercise: Sum the digits of an integer
# تمرین: جمع ارقام یک عدد صحیح
n = int(input("Enter a number: "))
sum_digits = 0
while n > 0:
    sum_digits += n % 10
    n //= 10
print("Sum of digits:", sum_digits)
```

```python
# Exercise: Calculate the average of 5 positive numbers entered by the user
# تمرین: محاسبه میانگین ۵ عدد مثبت وارد شده توسط کاربر
count = 0
total_sum = 0.0
while count < 5:
    n = int(input("Enter a positive number: "))
    if n < 0:
        print("Please enter a positive number.")
        continue
    total_sum += n
    count += 1
print("Average is:", total_sum / count)
```

```python
# Exercise: Find the largest digit in a number
# تمرین: پیدا کردن بزرگترین رقم در یک عدد
n = int(input("Enter a number: "))
max_digit = n % 10
while n > 0:
    if max_digit < n % 10:
        max_digit = n % 10
    n //= 10
print("Largest digit is:", max_digit)
```

```python
# تمرین: پیدا کردن کوچکترین رقم در یک عدد
n = int(input("Enter a number: "))
min_digit = n % 10
while n > 0:
    if min_digit > n % 10:
        min_digit = n % 10
    n //= 10
print("Smallest digit is:", min_digit)
```

```python
# Exercise: Menu-driven program for basic arithmetic operations
# تمرین: برنامه‌ای با منوی ساده برای عملیات حسابی پایه
while True:
    print("-------------------------------------")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")
    
    choice = int(input("Select an option: "))
    
    if choice == 5:
        print("Exiting the program.")
        break
    
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))

    if choice == 1:
        print("Result:", num1 + num2)
    elif choice == 2:
        print("Result:", num1 - num2)
    elif choice == 3:
        print("Result:", num1 * num2)
    elif choice == 4:
        if num2 == 0:
            print("Cannot divide by zero!")
        else:
            print("Result:", num1 / num2)
    else:
        print("Invalid option, please choose again.")
    
```
