تمرین: استفاده از colorama
یک برنامه بنویسید که یک لیست از رشته‌ها را دریافت کرده و هر رشته را با یک رنگ متفاوت در کنسول نمایش دهد. از کتابخانه‌ی colorama برای رنگ‌آمیزی متن‌ها استفاده کنید.

```python

strings = ["Hello", "World", "This", "Is", "Colorama"]
colors = [cl.Fore.RED, cl.Fore.GREEN, cl.Fore.YELLOW, cl.Fore.BLUE, cl.Fore.MAGENTA]

for string, color in zip(strings, colors):
    print(color + string)

print(cl.Style.RESET_ALL)
```

```python
import requests

# پارامترهای تاریخ شمسی
year = 1401
month = 11
day = 22

# آدرس API برای تاریخ شمسی
url = f'https://holidayapi.ir/jalali/{year}/{month}/{day}'

# ارسال درخواست GET
response = requests.get(url)

# بررسی موفقیت‌آمیز بودن پاسخ
if response.status_code == 200:
    data = response.json()
    print("آیا امروز تعطیل است؟", data['is_holiday'])
    print("مناسبت‌ها:")
    for event in data['events']:
        print(f"- {event['description']} (تعطیل: {event['is_holiday']})")
else:
    print(f"خطا در دریافت داده: {response.status_code}")
```

If you have a JSON string, you can parse it by using the json.loads() method.

```python
import json

# some JSON:
x =  '{ "name":"John", "age":30, "city":"New York"}'

# parse x:
y = json.loads(x)

# the result is a Python dictionary:
print(y["age"]) 
```

If you have a Python object, you can convert it into a JSON string by using the json.dumps() method.

```python
import json

# a Python object (dict):
x = {
  "name": "John",
  "age": 30,
  "city": "New York"
}

# convert into JSON:
y = json.dumps(x)

# the result is a JSON string:
print(y) 
```

You can convert Python objects of the following types, into JSON strings:

    dict
    list
    tuple
    string
    int
    float
    True
    False
    None


```python
import json

print(json.dumps({"name": "John", "age": 30}))
print(json.dumps(["apple", "bananas"]))
print(json.dumps(("apple", "bananas")))
print(json.dumps("hello"))
print(json.dumps(42))
print(json.dumps(31.76))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None)) 
```

```python
import json

x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

# use four indents to make it easier to read the result:
print(json.dumps(x, indent=4))
```

```python
import json

x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

# use . and a space to separate objects, and a space, a = and a space to separate keys from their values:
print(json.dumps(x, indent=4, separators=(". ", " = ")))
```

```python
import json

x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

# sort the result alphabetically by keys:
print(json.dumps(x, indent=4, sort_keys=True))
```
