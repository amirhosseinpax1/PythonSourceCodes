```python
class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal {self.name} created")

    def speak(self):
        print("Animal sound : ")

class Dog(Animal):
    def __init__(self, name, breed):
        Animal.__init__(self, name)  # فراخوانی __init__ کلاس والد بدون super()
        self.breed = breed
        print(f"Dog breed: {self.breed}")

    def speak(self):
        Animal.speak(self)  # فراخوانی speak کلاس والد بدون super()
        print("Bark!")

dog = Dog("Max", "Golden")
dog.speak()
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal {self.name} created")

    def speak(self):
        print("Animal sound : ")

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # فراخوانی __init__ کلاس والد
        self.breed = breed
        print(f"Dog breed: {self.breed}")

    def speak(self):
        super().speak()  # فراخوانی speak کلاس والد
        print("Bark!")

dog = Dog("Max", "Golden")
dog.speak()
```

Class Polymorphism

Polymorphism is often used in Class methods, where we can have multiple classes with the same method name.

For example, say we have three classes: Car, Boat, and Plane, and they all have a method called move():

```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Drive!")

class Boat:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Sail!")

class Plane:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang")       #Create a Car class
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat class
plane1 = Plane("Boeing", "747")     #Create a Plane class

for x in (car1, boat1, plane1):
  x.move()
```

Inheritance Class Polymorphism

```python
class Vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("drive!")

class Car(Vehicle):
  pass

class Boat(Vehicle):
  def move(self):
    print("Sail!")

class Plane(Vehicle):
  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang") #Create a Car object
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat object
plane1 = Plane("Boeing", "747") #Create a Plane object

for x in (car1, boat1, plane1):
  print(x.brand)
  print(x.model)
  x.move()
```

```python
# تعریف کلاس Book برای نگهداری اطلاعات کتاب‌ها
class Book:
    def __init__(self, title, author, copies):
        self.title = title       # عنوان کتاب
        self.author = author     # نویسنده کتاب
        self.copies = copies     # تعداد نسخه‌های موجود از کتاب

    def display_info(self):
        # نمایش اطلاعات کتاب
        print(f"Title: {self.title}, Author: {self.author}, Copies: {self.copies}")






# تعریف کلاس Member به عنوان پایه برای اعضای کتابخانه
class Member:
    def __init__(self, name, member_id):
        self.name = name                 # نام عضو
        self.member_id = member_id       # شناسه عضو
        self.borrowed_books = []         # لیست کتاب‌های امانت گرفته شده توسط عضو

    def borrow_book(self, book):
        # امانت گرفتن کتاب توسط عضو
        if book.copies > 0:
            book.copies -= 1            # کاهش تعداد نسخه‌های موجود
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed {book.title}")
        else:
            print(f"{book.title} is not available")

    def return_book(self, book):
        # بازگرداندن کتاب توسط عضو
        if book in self.borrowed_books:
            book.copies += 1            # افزایش تعداد نسخه‌های موجود
            self.borrowed_books.remove(book)
            print(f"{self.name} returned {book.title}")
        else:
            print(f"{self.name} did not borrow {book.title}")

# تعریف کلاس Student که از Member به ارث می‌برد
class Student(Member):
    def __init__(self, name, member_id, grade):
        super().__init__(name, member_id)  # استفاده از سازنده کلاس پایه
        self.grade = grade                # درجه تحصیلی دانشجو

    def borrow_book(self, book):
        # امانت گرفتن کتاب با محدودیت برای دانشجویان
        if len(self.borrowed_books) < 2:  # محدودیت امانت به ۲ کتاب
            super().borrow_book(book)    # فراخوانی متد borrow_book از کلاس پایه
        else:
            print(f"{self.name} cannot borrow more than 2 books")

# تعریف کلاس Teacher که از Member به ارث می‌برد
class Teacher(Member):
    def __init__(self, name, member_id, subject):
        super().__init__(name, member_id)  # استفاده از سازنده کلاس پایه
        self.subject = subject            # موضوع تدریس استاد

    def borrow_book(self, book):
        # امانت گرفتن کتاب با محدودیت برای اساتید
        if len(self.borrowed_books) < 5:  # محدودیت امانت به ۵ کتاب
            super().borrow_book(book)    # فراخوانی متد borrow_book از کلاس پایه
        else:
            print(f"{self.name} cannot borrow more than 5 books")

# تعریف کلاس Library برای مدیریت کتاب‌ها و اعضا
class Library:
    def __init__(self):
        self.__books = []   # لیست خصوصی از کتاب‌ها
        self.__members = [] # لیست خصوصی از اعضا

    def add_book(self, book):
        # افزودن کتاب به کتابخانه
        self.__books.append(book)
        print(f"Added {book.title} to the library")

    def add_member(self, member):
        # افزودن عضو به کتابخانه
        self.__members.append(member)
        print(f"Added {member.name} as a member")

    def display_books(self):
        # نمایش تمام کتاب‌های موجود در کتابخانه
        for book in self.__books:
            book.display_info()

# ایجاد کتاب‌ها
book1 = Book("Python Programming", "John Doe", 3)
book2 = Book("Data Science Handbook", "Jane Smith", 2)
book3 = Book()
# ایجاد اعضا
student1 = Student("Charlie", 201, "10th Grade")
teacher1 = Teacher("Dr. Smith", 301, "Mathematics")

# ایجاد کتابخانه و افزودن کتاب‌ها و اعضا
library = Library()
library.add_book(book1)
library.add_book(book2)
library.add_member(student1)
library.add_member(teacher1)

# نمایش کتاب‌های موجود در کتابخانه
library.display_books()

# امانت گرفتن و بازگرداندن کتاب توسط اعضا
student1.borrow_book(book1)  # خروجی: Charlie borrowed Python Programming
student1.borrow_book(book2)  # خروجی: Charlie borrowed Data Science Handbook
student1.borrow_book(book1)  # خروجی: Charlie cannot borrow more than 2 books

teacher1.borrow_book(book1)  # خروجی: Dr. Smith borrowed Python Programming
teacher1.borrow_book(book2)  # خروجی: Dr. Smith borrowed Data Science Handbook
teacher1.borrow_book(book1)  # خروجی: Dr. Smith borrowed Python Programming
```
